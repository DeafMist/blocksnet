```python
import blocksnet as bn

city = bn.City.from_pickle('new_model.pickle')
```

```python
bs = bn.BlocksSplitter(
  blocks=city.get_blocks_gdf(), 
  buildings=city.get_buildings_gdf().reset_index()
)
```

```python
blocks = bs.split_blocks()
```

```python
blocks.plot(figsize=(10,10)).set_axis_off()
```
