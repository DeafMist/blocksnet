# Generating intermodal graph and calculating adjacency matrix

```python
import os
import geopandas as gpd

local_crs = 32636
example_data_path = "./data"
```

## GraphGenerator

```python
from blocksnet import GraphGenerator

blocks = gpd.read_parquet(os.path.join(example_data_path, 'blocks.parquet')).to_crs(local_crs)
gg = GraphGenerator(blocks)
intermodal_graph = gg.run('intermodal')
```

```python
GraphGenerator.plot(intermodal_graph)
```

## AdjacencyCalculator

```python
from blocksnet import AdjacencyCalculator

ac = AdjacencyCalculator(blocks, intermodal_graph)
adj_mx = ac.run()
adj_mx.head()
```

```python
adj_mx.to_pickle(os.path.join(example_data_path, 'adjacency_matrix.pickle'))
```
